```python
#trying our hands on a simple mathematical code: solving a sorting problem with straight insertion 
vec = np.random.randint(0,100, size=5)
print(vec)
```

```python
#printing the elements of the vec array in ascending order
for j in np.arange(1, vec.size):
    v = vec[j]
    i=j
    while i>0 and vec[i-1] > v:
        vec[i] = vec[i-1]
        i= i-1
    vec[i]= v

print(vec)
```

```python
#printing the elements of the vec array in descending order
for j in np.arange(1, vec.size):
    v = vec[j]
    i=j
    while i>0 and vec[i-1] < v:
        vec[i] = vec[i-1]
        i= i-1
    vec[i]= v

print(vec)
```

```python
#Reading Data With Pandas: you can read different kind of files (excel, csv, json..)
df = pd.read_csv("\\Users\Delll\Downloads\daiabet.csv")
df.head()
```

```python
#Activity two:working with data problems
#first, lets check for any data problem
df.isna().sum()
```

there are no null values in this data

```python
#creating a list with numeric columns
columns= df.columns
categorical_columns = ['Sex', 'Complications']
columns = list(set(columns) - set(categorical_columns))
columns
```

```python
#using lambda function to replace 'no problem' with 'NaN' in the Complications 'column'
df['Complications'] = df['Complications'].apply(lambda x: np.nan if x=='no problem' else x)
```

```python
df.head(15)
```

```python
#checking data types
df.dtypes
```

```python
#converting the column 'Blood Sugar' into a float
df['blood_sugar'] = pd.to_numeric(df['blood_sugar'])
```

```python
#Converting non numeric columns to category type
df['Sex']= df['Sex'].astype('category')
df['Complications'] = df['Complications'].astype('category')
```

```python
#checking the non-numeric columns after conversion
df.dtypes
```

```python
#grouping the data by sex
df.groupby('Sex').sum().head()
```

```python
#Converting non numeric columns to category type
df['Sex']= df['Sex'].astype('object')
df['Complications'] = df['Complications'].astype('object')
```
