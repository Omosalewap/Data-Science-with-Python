```python
#firstly import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#let's plot an analytical function
x = np.linspace(-50, 50, 100)
y = np.power(x,2)
plt.plot(x,y)
```

```python
#similar analytical function but with different y-values
x = np.linspace(-50, 50, 100)
y = np.power(x,3)
plt.plot(x,y)
```

```python
#let's create a simple graph
fig, ax = plt.subplots()
x = np.linspace(0, 100, 500)
y = np.sin(2*np.pi*x/100)
ax.plot(x,y)
```

```python
#Creating a graph for a matheatical function
x = np.linspace(0,100,200)
y = np.sin(x)
```

```python
#Creating a subplot
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
#Creating line graph with OOA and panda dataframe using the data provided in the link below
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"

#reading our cvs data and viewing the first 5 rows
df = pd.read_csv(url)
df.head()
```

```python
#assigning column names to each column
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin', 'name']
df = pd.read_csv(url, names= column_names, delim_whitespace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#Checking for null values
df.isna().sum()
```

```python
#converting the columns 'horsepower' and 'year' to float and interger respectively
#column 'year' is already an interger but we'll convert to datetime
df.loc[df.horsepower == '?', 'horsepower'] = np.nan
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['year'] = pd.to_datetime(df.year, format='%y')
df['year'] = df['year'].dt.year
```

```python
df.info()
```

```python
df
```

```python
#plotting the graph of average horsepower per year
#we begin by grouping the average horsepower per year and assiging it to a variable 'avgh'
avgh= df.groupby('year')['horsepower'].mean()
avgh
```

```python
avgh.plot()
```

```python
#simpler method of ploting this graph
df.groupby('year')['horsepower'].mean().plot()
```

```python
#plotting the graph of average mpg per year
df.groupby('year')['mpg'].mean().plot()
```

```python
#plotting the graph of average displacement per year
df.groupby('year')['displacement'].mean().plot()
```

```python
#plotting the graph of average weight per year
df.groupby('year')['weight'].mean().plot()
```

```python
#plotting the count of cars produced per year
df.groupby('year')['name'].count().plot()
```
